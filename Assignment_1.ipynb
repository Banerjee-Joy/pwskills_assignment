**Q15. How to install Python on Windows and set path varibles?**

Ans - There are a few steps to install Python in Windows and set path variables. They are - 
        -> Firstly we have to download the desired version of Python executable installer for Windows from the official website of Python.
        -> Then we have to run the executable installer and install Python on Windows.
        -> After that we have to find the location where Python is installed in system and copy that location path from top by clicking over it.
        -> Next we have to open control panel, go to advanced system settings and open environment variables from there.
        -> Then we have to select and open path variables of user variables.
        -> After that we have to create a new path variable and paste the location path of the Python, that was copied previously.
        -> Now we have to open CMD and type Python there. If the environment variable is set then the Python command will run.
